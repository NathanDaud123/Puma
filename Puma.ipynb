```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def exploitation(sol, lb, ub, dim, n_sol, best, max_iter, iter, cost_function):
    Q = 0.67
    Beta = 2
    new_sol = np.copy(sol)
    sol_cost = np.array([cost_function(ind) for ind in sol])

    for i in range(n_sol):
        beta1 = 2 * np.random.rand()
        beta2 = np.random.randn(dim)
        w = np.random.randn(dim)
        v = np.random.randn(dim)
        F1 = np.random.randn(dim) * np.exp(2 - iter * (2 / max_iter))
        F2 = w * v**2 * np.cos(2 * np.random.rand() * w)
        mbest = np.mean(sol, axis=0)
        R_1 = 2 * np.random.rand() - 1
        S1 = (2 * np.random.rand() - 1 + np.random.randn(dim))
        S2 = (F1 * R_1 * sol[i] + F2 * (1 - R_1) * best)
        VEC = S2 / S1
        if np.random.rand() <= 0.5:
            Xatack = VEC
            if np.random.rand() > Q:
                new_sol[i] = best + beta1 * (np.exp(beta2) * (sol[np.random.randint(n_sol)] - sol[i]))
            else:
                new_sol[i] = beta1 * Xatack - best
        else:
            r1 = np.random.randint(n_sol)
            new_sol[i] = (mbest * sol[r1] - ((-1)**np.random.randint(2)) * sol[i]) / (1 + (Beta * np.random.rand()))

        new_sol[i] = np.maximum(new_sol[i], lb)
        new_sol[i] = np.minimum(new_sol[i], ub)
        new_sol_cost = cost_function(new_sol[i])

        if new_sol_cost < sol_cost[i]:
            sol[i] = new_sol[i]
            sol_cost[i] = new_sol_cost 

    return sol, sol_cost

def exploration(sol, lb, ub, dim, n_sol, cost_function):
    sol_cost = np.array([cost_function(ind) for ind in sol])
    sind = np.argsort(sol_cost)
    sol = sol[sind]
    sol_cost = sol_cost[sind]
    pCR = 0.20
    PCR = 1 - pCR
    p = PCR / n_sol

    for i in range(n_sol):
        x = sol[i]
        A = np.random.permutation(n_sol)
        A = A[A != i]
        a, b, c, d, e, f = A[:6]
        G = 2 * np.random.rand() - 1

        if np.random.rand() < 0.5:
            y = np.random.rand(dim) * (ub - lb) + lb
        else:
            y = sol[a] + G * (sol[a] - sol[b]) + G * (((sol[a] - sol[b]) - (sol[c] - sol[d])) + ((sol[c] - sol[d]) - (sol[e] - sol[f])))

        y = np.maximum(y, lb)
        y = np.minimum(y, ub)

        z = np.zeros_like(x)
        j0 = np.random.randint(dim)
        for j in range(dim):
            if j == j0 or np.random.rand() <= pCR:
                z[j] = y[j]
            else:
                z[j] = x[j]

        new_sol_cost = cost_function(z)
        
        if new_sol_cost < sol_cost[i]:
            sol[i] = z
            sol_cost[i] = new_sol_cost
        else:
            pCR += p

    return sol, sol_cost

def boundary_check(X, lb, ub):
    # Menggunakan operasi vektor NumPy untuk memeriksa batas dan mengganti nilai
    X = np.where(X > ub, ub, X)
    X = np.where(X < lb, lb, X)
    return X

# puma optimizer algorithm
def puma(n_sol, max_iter, lb, ub, dim, cost_function):

    # Parameter setting
    unselected = np.ones(2)  # 1:Exploration 2:Exploitation
    f3_explore = 0
    f3_exploit = 0
    seq_time_explore = np.ones(3)
    seq_time_exploit = np.ones(3)
    seq_cost_explore = np.ones(3)
    seq_cost_exploit = np.ones(3)
    score_explore = 0
    score_exploit = 0
    pf = [0.5, 0.5, 0.3]  # Probability factors
    pf_f3 = []
    mega_explore = 0.99
    mega_exploit = 0.99

    # Inisialisasi
    sol = np.random.uniform(lb, ub, (n_sol, dim))
    sol_cost = np.array([cost_function(ind) for ind in sol])
    best_index = np.argmin(sol_cost)
    best = sol[best_index]
    best_cost = sol_cost[best_index]
    convergence_individual = []
    convergence_cost = []

    # Fase Unexperienced
    for iter in range(3):
        # Eksplorasi
        sol_explor, sol_explor_cost = exploration(sol, lb, ub, dim, n_sol, cost_function)
        # Eksploitasi
        sol_exploit, sol_exploit_cost = exploitation(sol, lb, ub, dim, n_sol, best, max_iter, iter, cost_function)

        # Update solusi
        sol = np.concatenate((sol, sol_explor, sol_exploit), axis=0)
        sol_cost = np.concatenate((sol_cost, sol_explor_cost, sol_exploit_cost), axis=0)

        # Pemilihan solusi terbaik
        best_index = np.argmin(sol_cost)
        best = sol[best_index]
        best_cost = sol_cost[best_index]
        # print(best_cost)
        # print(sol_explor_cost)
        convergence_individual.append(best)
        convergence_cost.append(best_cost)
        print(f'Iteration: {iter+1} Best Cost = {best_cost}')

        # Hyper Initialization dan perhitungan skor
        seq_cost_explore[iter] = abs(best_cost - np.min(sol_explor_cost)) if iter == 0 else abs(np.min(sol_explor_cost) - seq_cost_explore[iter-1])
        seq_cost_exploit[iter] = abs(best_cost - np.min(sol_exploit_cost)) if iter == 0 else abs(np.min(sol_exploit_cost) - seq_cost_exploit[iter-1])

        if seq_cost_explore[iter] != 0:
            pf_f3.append(seq_cost_explore[iter])
        if seq_cost_exploit[iter] != 0:
            pf_f3.append(seq_cost_exploit[iter])
        
    # Perhitungan faktor dan skor
    f1_explore = pf[0] * (seq_cost_explore[0] / seq_time_explore[0])
    f1_exploit = pf[0] * (seq_cost_exploit[0] / seq_time_exploit[0])
    f2_explore = pf[1] * (np.sum(seq_cost_explore) / np.sum(seq_time_explore))
    f2_exploit = pf[1] * (np.sum(seq_cost_exploit) / np.sum(seq_time_exploit))

    score_explore = (pf[0] * f1_explore) + (pf[1] * f2_explore)
    score_exploit = (pf[0] * f1_exploit) + (pf[1] * f2_exploit)

    # Fase Experienced
    flag_change = 1
    for iter in range(3, max_iter):
        # Implementasi logika untuk fase experienced
        # Ini melibatkan pemilihan antara eksplorasi atau eksploitasi
        # berdasarkan skor yang dihitung dari hasil iterasi sebelumnya.
        if score_explore > score_exploit:
            # Eksplorasi
            select_flag = 1
            sol, sol_cost = exploration(sol, lb, ub, dim, n_sol, cost_function)
            unselected[1] += 1
            unselected[0] = 1
            f3_explore = pf[2]
            f3_exploit += pf[2]
            t_best_index = np.argmin(sol_cost)
            t_best = sol[t_best_index]
            t_best_cost = sol_cost[t_best_index]
            seq_cost_explore = np.roll(seq_cost_explore, 1)
            seq_cost_explore[0] = abs(best_cost - t_best_cost)
            if seq_cost_explore[0] != 0:
                pf_f3.append(seq_cost_explore[0])
            if t_best_cost < best_cost:
                best = t_best
                best_cost = t_best_cost
        else:
            # Eksploitasi
            select_flag = 2
            sol, sol_cost = exploitation(sol, lb, ub, dim, n_sol, best, max_iter, iter, cost_function)
            unselected[0] += 1
            unselected[1] = 1
            f3_explore += pf[2]
            f3_exploit = pf[2]
            t_best_index = np.argmin(sol_cost)
            t_best = sol[t_best_index]
            t_best_cost = sol_cost[t_best_index]
            seq_cost_exploit = np.roll(seq_cost_exploit, 1)
            seq_cost_exploit[0] = abs(best_cost - t_best_cost)
            if seq_cost_exploit[0] != 0:
                pf_f3.append(seq_cost_exploit[0])
            if t_best_cost < best_cost:
                best = t_best
                best_cost = t_best_cost

        if flag_change != select_flag:
            flag_change = select_flag
            seq_time_explore = np.roll(seq_time_explore, 1)
            seq_time_explore[0] = unselected[0]
            seq_time_exploit = np.roll(seq_time_exploit, 1)
            seq_time_exploit[0] = unselected[1]

        # Hyper Initialization dan perhitungan skor
        f1_explore = pf[0] * (seq_cost_explore[0] / seq_time_explore[0])
        f1_exploit = pf[0] * (seq_cost_exploit[0] / seq_time_exploit[0])
        f2_explore = pf[1] * (np.sum(seq_cost_explore) / np.sum(seq_time_explore))
        f2_exploit = pf[1] * (np.sum(seq_cost_exploit) / np.sum(seq_time_exploit))

        # Penyesuaian nilai Mega
        if score_explore < score_exploit:
            mega_explore = max((mega_explore - 0.01), 0.01)
            mega_exploit = 0.99
        elif score_explore > score_exploit:
            mega_explore = 0.99
            mega_exploit = max((mega_exploit - 0.01), 0.01)

        lmn_explore = 1 - mega_explore
        lmn_exploit = 1 - mega_exploit

        # Perhitungan skor baru
        score_explore = (mega_explore * f1_explore) + (mega_explore * f2_explore) + (lmn_explore * (min(pf_f3) * f3_explore))
        score_exploit = (mega_exploit * f1_exploit) + (mega_exploit * f2_exploit) + (lmn_exploit * (min(pf_f3) * f3_exploit))

        # Update convergence_cost history
        convergence_individual.append(best)
        convergence_cost.append(best_cost)
        print(f'Iteration: {iter+1} Best Cost = {best_cost}')

    unique_values, first_occurrence_indexes = np.unique(convergence_cost, return_index=True)
    last_three_unique_values = unique_values[0:3]
    last_three_indexes = first_occurrence_indexes[0:3]
    
    i = 0
    for individual in last_three_indexes:
        print("Indeks dari 3 Nilai Unik Terakhir:", individual)
        print("3 Nilai Unik Terakhir:", last_three_unique_values[i])
        daftar_makanan(convergence_individual[individual])
        i = i + 1
        
    return best, best_cost, convergence_cost
```
